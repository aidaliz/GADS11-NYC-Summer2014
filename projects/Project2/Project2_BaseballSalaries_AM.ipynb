```python
import numpy as np
import pandas as pd
import pylab as pl
```

```python
#Read the data
batting_file_path = "/users/aidalizmaldonado/GADS11-NYC-Summer2014/projects/Project2/baseball.csv"
pitching_file_path = "/users/aidalizmaldonado/GADS11-NYC-Summer2014/projects/Project2/pitching.csv"
#batting_data = pd.DataFrame.from_csv(batting_file_path, index_col = None)
#pitching_data = pd.DataFrame.from_csv(pitching_file_path, index_col = None)
batting_data = pd.read_csv(batting_file_path)
pitching_data = pd.read_csv(pitching_file_path)
#batting_data.head(10)
#pitching_data.head(10)
```

```python
#Drop columns that are not useful
batting_data = batting_data.drop(['lahmanID','nameGiven','managerID','hofID','birthMonth','birthDay','birthCountry','birthState',
'birthCity','deathYear','deathMonth','deathDay','deathCountry','deathState','deathCity','nameFirst','nameLast','nameNote','nameNick',
'debut','finalGame','college','lahman40ID','lahman45ID','retroID','holtzID','bbrefID','deathDate','birthDate','stint'], axis = 1)
pitching_data = pitching_data.drop(['stint','lgID','G'], axis = 1)
batting_data.head(10)
pitching_data.head(10)

joined_data = pd.merge(batting_data,pitching_data, on = ['playerID','yearID'])
joined_data.ix[0:10, ['playerID','yearID','G_batting','HR','ERA','birthYear']]
```

```python
#Add new batting features

#Create a dummy column for identifying players that are pitchers
joined_data['Pitcher']=0
joined_data['Pitcher'][joined_data['ERA'] > 0] = 1
joined_data.ix[0:10, ['playerID','yearID','G_batting','HR','ERA','Pitcher']]
joined_data['AGE'] = joined_data['yearID'] - joined_data['birthYear']

#Batting Average = hits/At Bats
joined_data['AVG'] = joined_data['H']/joined_data['AB']

#Slugging Average = [hits + 2*doubles + 3*tripples + 4*home runs]/At bats
joined_data['SA'] = (joined_data['H'] + 2*joined_data['X2B'] + 3*joined_data['X3B'] + 4*joined_data['HR'])/joined_data['AB']

#Strikeout ratio = Strikeouts/At Bats
joined_data['SR'] = joined_data['SO']/joined_data['AB']

#Stolen Bases ratio = Stolen Bases/(Stolent Bases + Caught Stealing)
joined_data['SBR'] = joined_data['SB']/(joined_data['SB']+joined_data['CS'])


#Add new features to the pitching dataset

#Pitcher Strikeout Ratio = Pitcher Strikeouts/Batters Faced by Pitcher
joined_data['PSR'] = joined_data['pitch_SO']/joined_data['BFP']

#Left on base ratio = (Hits Allowed + Walks Allowed + Batters hit by pitch - Runs Allowed)/(Hits Allowed + Walks Allowed + Batters hit by pitch + -(1.4*Home Runs Allowed))
joined_data['LOB'] = (joined_data['pitch_H'] + joined_data['pitch_BB'] + joined_data['pitch_HBP'] - joined_data['pitch_R'])/(joined_data['pitch_H'] + joined_data['pitch_BB'] + joined_data['pitch_HBP'] + joined_data['pitch_R'] - 1.4*joined_data['pitch_HR'])

#Winning ratio = Wins / (Wins + Losses)
joined_data['WR'] = joined_data['W']/(joined_data['W'] + joined_data['L'])
joined_data.ix[:, ['playerID','yearID','lgID','G','G_batting','HR','ERA','Pitcher','SA','AVG','SR', 'SBR','PSR','LOB','WR']]

#Additional ration to consider: cummulative games, hits, HR's, years with club, age, club
#pandas panel, multi-indices

# Replace Inf values with Nan
joined_data = joined_data.replace([inf, -inf], np.nan)
```
